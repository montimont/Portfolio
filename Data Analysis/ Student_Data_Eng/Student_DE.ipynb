```python
import pandas as pd
if 'transformer' not in globals():
    from mage_ai.data_preparation.decorators import transformer
if 'test' not in globals():
    from mage_ai.data_preparation.decorators import test


@transformer
def transform(df, *args, **kwargs):
    high_school_info = df[['high_school_name', 'high_school_state', 'high_school_type']]

    high_school_info = high_school_info.drop_duplicates(subset=['high_school_name']).reset_index(drop=True)

    high_school_info['high_school_id'] = high_school_info.index

    high_school_dim = high_school_info[['high_school_id', 'high_school_name', 'high_school_state', 'high_school_type']]

    university_info = df[['university_name', 'unversity_state', 'unversity_type']]

    university_school_info = university_info.drop_duplicates(subset=['university_name']).reset_index(drop=True)

    university_school_info['university_id'] = university_school_info.index

    university_dim = university_school_info[['university_id', 'university_name', 'unversity_state', 'unversity_type']]

    major_info = df[['university_major']]

    university_major_info = major_info.drop_duplicates(subset=['university_major']).reset_index(drop=True)

    university_major_info['major_id'] = university_major_info.index

    major_dim = university_major_info[['major_id', 'university_major']]

    university_dim = university_dim.rename(columns={'unversity_state': 'university_state', 'unversity_type': 'university_type'})

    student_table = df.merge(high_school_dim, on='high_school_name') \
                    .merge(university_dim, on='university_name') \
                    .merge(major_dim, on='university_major') \
                    [['student_id', 'student_full_name', 'address', 'state','high_school_id', 'university_id', 'major_id']]

    return {
        'high_school_dim': high_school_dim.to_dict(orient='dict'),
        'university_dim': university_dim.to_dict(orient='dict'),
        'major_dim': major_dim.to_dict(orient='dict'),
        'student_table': student_table.to_dict(orient='dict')
    }

@test
def test_output(output, *args) -> None:
    assert output is not None, 'The output is undefined'
```

###SQL

```python
--- Fix the Chicago Error ---
UPDATE student_data_engineering_port.university_dim
SET University_state = 'IL'
WHERE University_name = 'University of Illinois at Chicago'
AND (University_state = 'GA' OR University_state = 'HI' OR University_state = 'ID');

--- University Enrollment Analysis: Total Students by University Name and State ---

SELECT
    ud.university_name,
    ud.university_state,
    COUNT(st.student_id) AS total_students
FROM
    student-data-portfolio.student_data_engineering_port.student_table st
JOIN
    student-data-portfolio.student_data_engineering_port.university_dim ud
ON
  ud.university_id = st.university_id
GROUP BY
    ud.university_name,
    ud.university_state
ORDER BY
    total_students DESC;


--- University Major Enrollment Analysis: Student Percentage by University Major ---

WITH student_counts AS (
    SELECT
        major_dim.major_id,
        major_dim.university_major,
        COUNT(student_table.student_id) AS total_students
    FROM
        student_data_engineering_port.student_table AS student_table
    LEFT JOIN
        student_data_engineering_port.major_dim AS major_dim
    ON
        student_table.major_id = major_dim.major_id
    GROUP BY
        major_dim.major_id,
        major_dim.university_major
)
SELECT
    university_major,
    total_students,
    total_students / (SELECT SUM(total_students) FROM student_counts) AS student_percentage
FROM
    student_counts
ORDER BY
    student_percentage DESC;
```
